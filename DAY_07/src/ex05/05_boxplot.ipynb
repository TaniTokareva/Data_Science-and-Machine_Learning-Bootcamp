## Визуализация boxplot

Создадим фигуру (fig) с двумя подграфиками (ax) в одном ряду (1) и двумя столбцами (2).

Получаем объект fig для настройки всей фигуры и массив ax из двух осей (графиков), ax[0] и ax[1].

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8)) 
df[df['time'] == 'after'].boxplot(column='diff', 
                                by=['group'], 
                                  ax=ax[0], 
                                  color=dict(whiskers='red',
                                             medians='red', caps='red'),
                                  boxprops=dict(color = 'green', linewidth=3), 
                                  medianprops=dict(linewidth=2)) 
df[df['time'] == 'before'].boxplot(column='diff',
                                   by='group',
                                   ax=ax[1],
                                   color=dict(whiskers='red',
                                              medians='red', caps='red'),
                                   boxprops=dict(color = 'green', linewidth=3),
                                   medianprops=dict(linewidth=2))
ax[0].set_title('after', fontsize=15)
ax[1].set_title('before', fontsize=15)
fig.suptitle(
    'Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```

Проверка (для себя)

```python
# Фильтрация данных для контрольной группы и времени "before"
control_before = df[(df['group'] == 'control') & (df['time'] == 'before')]

q1 = control_before['diff'].quantile(0.25)  
q3 = control_before['diff'].quantile(0.75) 
iqr = q3 - q1 

print(iqr)
```

```python
control_after = df[(df['group'] == 'control') & (df['time'] == 'after')]

q1 = control_after['diff'].quantile(0.25)  
q3 = control_after['diff'].quantile(0.75) 
iqr = q3 - q1 

print(iqr)
```
