```python
import pandas as pd
```

загрузите файл JSON, созданный в предыдущем упражнении, в фрейм данных
установить CarNumber в качестве индексного столбца

```python
df = pd.read_json('../data/auto.json',
                 orient='records')
df.set_index('CarNumber', inplace=True)
df
```

## Выборки

### 1. отображать только те строки, где штрафы превышают 2100


```python
df[df['Fines'] > 2100]
```

### 2. отображать только те строки, где штрафы больше 2100, а возврат равен 2

```python
df[df['Fines'] > 2100 & (df['Refund'] == 2)]
```

### 3.отображать только те строки, где есть модели из списка:[’Focus’, ’Corolla’]

```python
df[df['Model'].isin(['Focus', 'Corolla'])]
```

отображать только те строки, где номер автомобиля есть из списка: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]

```python
df[df.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]
```

## Агрегации с маркой и моделью

### 1. display the median fines grouped by the make

```python
df.groupby(by=['Make']).agg({'Fines': ['median']})
```

### 2. display the median fines grouped by the make and the model

```python
df.groupby(by=['Make', 'Model']).agg({'Fines' : ['median']})
```

### 3. display the number of fines grouped by the make and the model in order to understand if we can trust the median values

```python
df.groupby(by=['Make', 'Model']).agg({'Fines': ['count']})
```

### 4.display the minimum and the maximum fines grouped by the make and the model in order to better understand the variance

```python
df.groupby(by=['Make', 'Model']).agg({'Fines' : ['min', 'max']})
```

5. display the standard deviation of the fines grouped by the make and the model in order to better understand the variance

```python
df.groupby(by=['Make', 'Model']).agg({'Fines' : ['std']})
```

## Агрегация по номерам машины

### 1. display the car numbers grouped by the number of the fines in descending order, we want to find those who most often violated the law

```python
df.groupby(by=['CarNumber']).agg({'Fines' : ['count']}).sort_values(by=('Fines', 'count'), ascending=False)
```

### 2. select from the initial dataframe all the rows corresponding to the top-1 car number, we want to zoom in a little bit

```python
df[df.index == 'Y7689C197RUS']
```

### 3. display the car numbers grouped by the sum of the fines in descending order, we want to find those who paid the most

```python
df.groupby(by='CarNumber').agg({'Fines' : ['sum']}).sort_values(by=('Fines', 'sum'), ascending=False)
```

### 4. select from the initial dataframe all the rows corresponding to the top-1 car number, we want to zoom in a little bit

```python
df[df.index == 'X758HY197RUS']
```

### 5. display a table that answers the question: are there any car numbers that were connected to different models?

```python
df.groupby(by=['CarNumber', 'Model']).agg({'Model': ['nunique']}).sort_values(by=('Model', 'nunique'), ascending=False)
```

Вывод: разных машин с одинаковыми номерами нет.
