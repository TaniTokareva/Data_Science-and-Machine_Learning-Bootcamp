Текст задания:
создать новую колонку дневное время
вам необходимо присвоить конкретное значение времени суток, если час находится в пределах определенного интервала, например, полдень, если час больше 11 и меньше или равен 17
0 – 3.59 ночь, 4 – 6.59 раннее утро, 7 – 10.59 утро, 11 – 16.59 день, 17 – 19.59 ранний вечер, 20 – 23.59 вечер
используйте метод cut для решения этой подзадачи
назначить столбец user в качестве индекса

```python
time_bins=[-1, 3, 6, 10, 16, 19, 23]
labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
df['daytime'] = pd.cut(df['hour'],
                      bins=time_bins,
                      labels=labels)
df.set_index('user', inplace=True)
df
```

## Минимум и максимум времени

```python
total_elements = df.shape[0]
time_of_day_counts = df["daytime"].value_counts()
sorted_df = df.sort_values(by=["hour", "minute", "second"])
min_hour = df["hour"].min()
max_hour = df["hour"].max()
print("Минимальное время - ", min_hour)
print("Максимальное время - ", max_hour)
```

```python
max_night = df[df['daytime'] == 'night']
max_night = max_night['hour'].max()
print("Максимальное ночное время - ", max_night)
```

```python
min_morning = df[df['daytime'] == 'morning']
min_morning = min_morning['hour'].min()
print("Минимальное утреннее время - ", min_morning)
```

## И кто же посещал страницу в такое время?

```python
df[df['hour'] == max_night]
```

```python
df[df['hour'] == min_morning]
```

## Мода категории "daytime" и "hour"

```python
mode_daytime = df["daytime"].mode()[0]
print(mode_daytime)
```

```python
mode_hour = df['hour'].mode()[0]
print(mode_hour)
```

## 3 самых ранних часа утра и соответствующие им имена пользователей, а также 3 самых поздних часа и имена пользователей с помощью nsmallest() и nlargest()

```python
early_morning_3 = df[df['daytime'] == 'morning'].nsmallest(n=3, columns='hour')
early_morning_3
```

```python
late_night_3 = df[df['daytime'] == 'night'].nlargest(n=3, columns='hour')
late_night_3
```
